This notebook provides code for visualizing the long-tail distribution property of the training dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

#  FungiTastic  CSV 
df = pd.read_csv("data/fungi-clef-2025/metadata/FungiTastic-FewShot/FungiTastic-FewShot-Train.csv")  # 

# 
class_counts = df['category_id'].value_counts().sort_values(ascending=False)

# 
plt.figure(figsize=(12, 4))
plt.bar(range(len(class_counts)), class_counts.values)
plt.xlabel('Class Index (Sorted)')
plt.ylabel('Number of Samples')
plt.title('Training Set Class Distribution (Long-Tailed)')
plt.tight_layout()
plt.show()

#  1~5 6~10 
bins = [0, 5, 10, 20, 50, 100]
plt.figure(figsize=(6, 4))
plt.hist(class_counts.values, bins=bins, edgecolor='black')
plt.xlabel('Number of Samples per Class')
plt.ylabel('Number of Classes')
plt.title('Histogram of Class Frequencies')
plt.tight_layout()
plt.show()

# 
summary = {
    "Total classes": len(class_counts),
    "Classes with 5 samples": (class_counts <= 5).sum(),
    "Median samples per class": int(class_counts.median()),
    "Max samples in a class": int(class_counts.max()),
    "Min samples in a class": int(class_counts.min()),
}
summary
```
